## DF_Final (Item lvl detail)

```python
# df_final = pd.concat([df,df_zip],axis=1)df_final['when_the_Jumpman_arrived_at_dropoff']
# df_final = pd.read_pickle('data/df_final.pickle')

def df_final_clean(df):
    df['total_time_to_deliver'] = df['time_to_deliver']
    df = df.drop('time_to_deliver',axis=1)
    df['time_start_to_pickup'] = (df['when_the_Jumpman_arrived_at_pickup'].\
                               sub(df['when_the_delivery_started'])).astype('timedelta64[m]')
    df['time_pickup_depart_to_dropoff'] = (df['when_the_Jumpman_arrived_at_dropoff'].\
                               sub(df['when_the_Jumpman_left_pickup'])).astype('timedelta64[m]')
    df['time_pickup_arrival_to_depart'] = (df['when_the_Jumpman_left_pickup'].\
                               sub(df['when_the_Jumpman_arrived_at_pickup'])).astype('timedelta64[m]')
    df['item_quantity'] = df['item_quantity'].replace(np.nan,1)
    df['day'] = df['when_the_Jumpman_arrived_at_dropoff'].dt.day
    df['weekday'] = df['when_the_Jumpman_arrived_at_dropoff'].dt.day_name()
    df = df[df_final_cols]
    return df
```
